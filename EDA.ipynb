Bivariate Analysis

```python
print(df.groupby('Digital_Eye_Strain_Risk')['Daily_Screen_Hours'].mean())
```

```python
## Do users with eye strain spend more time on screens?
plt.figure(figsize=(4,2))
sns.set_style("whitegrid")
print(f"Do users with eye strain spend more time on screens? \n")
y = df.groupby('Digital_Eye_Strain_Risk')['Daily_Screen_Hours'].mean()
plt.bar(y.index,y.values)
plt.title("eye strain vs avg screen hr")
plt.xlabel('Eye strain risk')
plt.ylabel("Mean Screen Time")
plt.tight_layout()
plt.show()
```

```python
print(f"Which age group has the highest screen exposure? \n")
plt.figure(figsize = (7,3.5 ))
y = df.groupby('Age')['Daily_Screen_Hours'].mean()
plt.bar(y.index, y.values)
plt.xlabel("Age")
plt.ylabel("avg value")
plt.title("Age vs avg screen hr")
plt.tight_layout()
plt.show()
```
